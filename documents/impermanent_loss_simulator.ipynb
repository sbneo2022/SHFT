```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def iloss_simulate(base_pct_chg=0, quote_pct_chg=0):
    """Calculate simulated impermanent loss from an initial value invested,
    Args:
        base_pct_chg (float): percentage change for first token of LP pair
        quote_pct_chg (float): percentage change for second token of LP pair
    
    Returns:
        tuple (value_f, iloss): final value of value invested, and decimal impermanent loss
    """

    ratio = (1 + base_pct_chg) / (1 + quote_pct_chg)
    return 2 * (ratio ** 0.5 / (1 + ratio)) - 1

```

```python
iloss_simulate(+.35, -.2)
```

```python
x = np.linspace(-.5, .5, 200)
y = np.linspace(-.5, .5, 200)

X, Y = np.meshgrid(x, y)
Z = iloss_simulate(X, Y)
```

```python
fig = plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 1000, cmap='binary')
ax.set_xlabel('Percentage of variation of price for asset X')
ax.set_ylabel('Percentage of variation of price for asset Y')
ax.set_zlabel('Impermanent loss suffered');
ax.view_init(30, 25)
```
